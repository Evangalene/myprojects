```python
# Clean actual_price and discounted_price
df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
df['discounted_price'] = df['discounted_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Filter data between ₹500 and ₹3000
filtered_df = df[(df['actual_price'] >= 500) & (df['actual_price'] <= 3000)].copy()

# Step 2: Bin actual prices
bins = [500, 1000, 2000, 3000]
labels = ['Low (500-1000)', 'Mid (1000-2000)', 'High (2000-3000)']
filtered_df['price_range'] = pd.cut(filtered_df['actual_price'], bins=bins, labels=labels, include_lowest=True)

# Step 3: Group and calculate average discounted price
avg_discount_by_range = filtered_df.groupby('price_range')['discounted_price'].mean().reset_index()

# Step 4: Bar Plot
plt.figure(figsize=(8, 5))
sns.barplot(x='price_range', y='discounted_price', data=avg_discount_by_range, palette='viridis')
plt.title("Average Discounted Price by Actual Price Range (₹500–₹3000)")
plt.xlabel("Actual Price Range")
plt.ylabel("Average Discounted Price (₹)")
plt.grid(True)
plt.show()
```

```python
plt.plot(df['discounted_price'].head(100))
plt.title("Discounted Price Trend (First 100 Products)")
plt.xlabel("Index")
plt.ylabel("Discounted Price (₹)")
plt.grid(True)
plt.show()
```

```python
#Conclusions
highest_rating_row = df.loc[df['rating'].idxmax()]
highest_rating_product = highest_rating_row['product_name']
highest_rating_value = highest_rating_row['rating']
highest_rating_count_row = df.loc[df['rating_count'].idxmax()]

# Print conclusions
print(f"👉 Product with the highest rating count: {highest_rating_count_row['product_name']}")

highest_price_row = df.loc[df['actual_price'].idxmax()]
print(f"👉 Highest price product: {highest_price_row['product_name']} with price ₹{highest_price_row['actual_price']}")

print(f"👉 Product with highest rating: {highest_rating_product} with rating {highest_rating_value}")
```
