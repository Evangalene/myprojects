```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("iris_csv.csv")
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
sns.lineplot(x='sepallength',y='sepalwidth',data=df)
plt.show()
```

```python
sns.countplot(x='class',data=df)
plt.show()
```

```python
sns.scatterplot(x='sepallength',y='sepalwidth',hue='class',data=df)
plt.legend(bbox_to_anchor=(1,1),loc=2)
plt.show()
```

```python
def graph(y):
    sns.boxplot(x="class",y=y,data=df)
plt.figure(figsize=(10,10))
plt.subplot(221)
graph('sepallength')
plt.subplot(222)
graph('sepalwidth')
plt.subplot(223)
graph('petallength')
plt.subplot(224)
graph('petalwidth')
plt.show()
```

```python
plt.boxplot(x='sepallength',data=df)
plt.show()
```

```python
numerical_df = df.select_dtypes(include=['float64', 'int64'])
x=numerical_df.corr()
sns.heatmap(x,cmap='coolwarm',annot=True)
plt.show()
```

```python
plt.hist(df['sepallength'],bins=10)
plt.show()
```
