```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
data={'c1':[1,3,7,8],'c2':[3,6,5,2],'c3':[5,2,1,3]}
df=pd.DataFrame(data)
sc=preprocessing.MinMaxScaler(feature_range=(0,1))
ds=sc.fit_transform(df)
print("scaled data",ds)
      
```

```python
nd=preprocessing.normalize(df,norm='l1')
print("normalized data",nd.round(2))
```

```python
bd=preprocessing.Binarizer(threshold=5).transform(df).round(2)
print(bd)
```

```python
from scipy.stats import tmean
x_train=np.array([[1,1,2],[3,2,1],[4,2,3]])
print("mean",tmean(x_train).round(2))
print("sd",round(x_train.std(),2))
```

```python
from sklearn.preprocessing import scale
sx=scale(x_train)
print("standardized data",sx.round(2))
print("mean",tmean(sx).round(2))
print("std",round(sx.std(),2))
```

```python
import matplotlib.pyplot as plt
x=[24,35,67,88]
y=[44,55,62,90]
rng=np.random.RandomState(0)
colors=rng.rand(4)
size=1000*rng.rand(4)
plt.scatter(x,y,color=colors,s=size,marker='o')
plt.title("bar")
plt.show()
```

```python
labels='apple','banana','mango','orange'
colors=['red','yellow','green','orange']
sizes=[20,34,52,62]
explode=[0,0,0,1]
plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%",explode=explode,shadow=True)
plt.title("pie")
plt.axis('equal')
```
